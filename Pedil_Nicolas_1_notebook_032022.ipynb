```python
import numpy as np #On importe les bibliothèques correspondantes à savoir Numpy, Pandas et SeaBorn
import pandas as pd
import seaborn as sns
```

```python
aide_alimentaire = pd.read_csv("aide_alimentaire.csv") #On importe les fichiers CSV correspondants à savoir aide_alimentaire,...
dispo_alimentaire = pd.read_csv("dispo_alimentaire.csv") #dispo_alimentaire,...
population = pd.read_csv("population.csv") #population,...
sous_nutrition = pd.read_csv("sous_nutrition.csv") #et sous_nutrition.
```

```python
#On prépare les données du tableau Population 
population['Valeur'] = population['Valeur'] * 1000 # On convertit la population exprimée en milliers en unités
population["Année"] = pd.to_numeric(population["Année"], downcast="integer") # On convertit les années en integer
population["Valeur"] = pd.to_numeric(population["Valeur"], downcast="integer") #On convertit les populations en integer
population.rename(columns={"Valeur":"Population"}, inplace=True) #On renomme la colonne "Valeur" en "Population"
population.head() #On jette un coup d'oeil à population
```

```python
#On prépare les données du tableau Sous-nutrition
sous_nutrition.fillna(value={"Valeur": 0}, inplace=True) # On remplace les valeurs inconnues par des 0
sous_nutrition['Valeur'] = sous_nutrition['Valeur'].str.replace('<','') #On remplace les "<" par rien
sous_nutrition['Valeur'] = sous_nutrition['Valeur'].str.replace('0.1','0.0') #On remplace 100 000 par 0
sous_nutrition["Valeur"] = pd.to_numeric(sous_nutrition["Valeur"], downcast="float") #On convertit en float
sous_nutrition['Valeur'] = sous_nutrition['Valeur'] * 1000000 # On convertit la population exprimée en millions en unités
sous_nutrition['Année'] = sous_nutrition['Année'].str.replace('2012-2014','2013') #On convertit les intervalles en moyenne
sous_nutrition['Année'] = sous_nutrition['Année'].str.replace('2013-2015','2014')
sous_nutrition['Année'] = sous_nutrition['Année'].str.replace('2014-2016','2015')
sous_nutrition['Année'] = sous_nutrition['Année'].str.replace('2015-2017','2016')
sous_nutrition['Année'] = sous_nutrition['Année'].str.replace('2016-2018','2017')
sous_nutrition['Année'] = sous_nutrition['Année'].str.replace('2017-2019','2018')
sous_nutrition["Année"] = pd.to_numeric(sous_nutrition["Année"], downcast="integer") #Puis on convertit en entier les années
sous_nutrition.rename(columns={"Valeur":"Personnes en sous-nutrition"}, inplace=True) #On fait de même avec sous-nutrition
sous_nutrition.head() #On affiche le DataFrame sous-nutrition
```

```python
nutripop = pd.merge(population, sous_nutrition) #On joint sous_nutrition à population
nutripop_mask=nutripop['Année']==2017 #On filtre le dataframe nutripop pour ne garder que les données de 2017
filtered_nutripop = nutripop[nutripop_mask]
filtered_nutripop.head()
```

```python
filtered_nutripop.fillna(0, inplace=True) #On remplace les valeurs inconnues par des 0
filtered_nutripop.head()
```

```python
poptotal=filtered_nutripop["Population"].sum() # On somme les populations de chaque pays
sousnutritotal=filtered_nutripop["Personnes en sous-nutrition"].sum() # On somme les nombres de personnes en sn dans chaque pays
print("La proportion de personnes en sous-nutrition est de", round((sousnutritotal/poptotal)*100,2) , "%.")
```

```python
filtered_nutripop['Proportion en sn'] = filtered_nutripop['Personnes en sous-nutrition']/filtered_nutripop['Population']
# On calcule la proportion pour chaque pays des personnes en sous-nutrition
filtered_nutripop['Proportion en sn'] = filtered_nutripop['Proportion en sn']*100 # On la met en pourcentage
top10sn = filtered_nutripop.sort_values(by = 'Proportion en sn', ascending = False) # On classe par proportion décroissante, 
# pour avoir les pays les plus touchés c'est-à-dire ceux ayant la plus grande proportion de personnes en sous-nutrition
top10sn.head(10) #on affiche les 10 premiers
```

```python
top10snabs = filtered_nutripop.sort_values(by = 'Personnes en sous-nutrition', ascending = False) # On classe  
# pour avoir les pays les plus touchés en valeur absolue
top10snabs.head(10) #on affiche les 10 premiers
```

```python
dispo_alimentaire.fillna(0, inplace=True) #On remplace les valeurs inconnues par des 0
dispo_alimentaire.head()
```

```python
dispoadd = dispo_alimentaire [['Zone', 'Disponibilité alimentaire (Kcal/personne/jour)']] # On ne garde que les colonnes utiles
dispoadd = dispoadd.groupby(['Zone'], as_index = False) #On groupe par pays la dispo alim en gardant la colonne Zone
dispoadd = dispoadd.sum()
dispoaddm =  pd.merge(dispoadd, filtered_nutripop) #On joint disponibilité alimentaire à données de la population en 2017
dispoaddm['Disponibilité alimentaire du pays'] = dispoaddm['Disponibilité alimentaire (Kcal/personne/jour)']*dispoaddm['Population']
dispoaddm.head()
nombrepers = dispoaddm['Disponibilité alimentaire du pays'].sum()/2250
print("Le nombre de personnes pouvant théoriquement être nourries est de", round(nombrepers), "personnes.")
```

```python
dispo_alimentaire2 = dispo_alimentaire[dispo_alimentaire['Origine'] == 'vegetale'] #On ne garde que les aliments ayant une origine végétale
dispoadd2 = dispo_alimentaire2 [['Zone', 'Disponibilité alimentaire (Kcal/personne/jour)']] #On groupe par pays la dispo alim 
dispoadd2 = dispoadd2.groupby(['Zone'], as_index = False)
dispoadd2 = dispoadd2.sum()
dispoaddm2 =  pd.merge(dispoadd2, filtered_nutripop) #On joint disponibilité alimentaire à données de la population en 2017 
dispoaddm2['Disponibilité alimentaire du pays'] = dispoaddm2['Disponibilité alimentaire (Kcal/personne/jour)']*dispoaddm2['Population']
nombrepers2 = dispoaddm2['Disponibilité alimentaire du pays'].sum()/2250
print("Le nombre de personnes pouvant théoriquement être nourries avec des végétaux est de", round(nombrepers2), "personnes.")
```

```python
#On calcule les 3 proportions demandées : l’utilisation de la disponibilité intérieure 
# en particulier la part qui est attribuée à l’alimentation animale, 
# celle qui est perdue et celle qui est concrètement utilisée pour l'alimentation humaine. 
Prop_animale = (dispo_alimentaire["Aliments pour animaux"].sum() / dispo_alimentaire["Disponibilité intérieure"].sum())*100
Prop_perdue = (dispo_alimentaire["Pertes"].sum() / dispo_alimentaire["Disponibilité intérieure"].sum())*100
Prop_humaine = (dispo_alimentaire["Nourriture"].sum() / dispo_alimentaire["Disponibilité intérieure"].sum())*100
Prop_sem = (dispo_alimentaire["Semences"].sum() / dispo_alimentaire["Disponibilité intérieure"].sum())*100
Prop_traite = (dispo_alimentaire["Traitement"].sum() / dispo_alimentaire["Disponibilité intérieure"].sum())*100
Prop_autres = (dispo_alimentaire["Autres Utilisations"].sum() / dispo_alimentaire["Disponibilité intérieure"].sum())*100
print("La proportion est de", round(Prop_animale), "% pour la part attribuée à l'alimentation animale.")
print("La proportion est de", round(Prop_perdue), "% pour la part qui est perdue.")
print("La proportion est de", round(Prop_humaine), "% pour la part attribuée à l'alimentation humaine.")
print("La proportion est de", round(Prop_sem), "% pour la part attribuée à la semence.")
print("La proportion est de", round(Prop_traite), "% pour la part attribuée au traitement.")
print("La proportion est de", round(Prop_autres), "% pour la part attribuée aux autres utilisations.")
```

```python
#On ne garde que les données concernant les céréales
cereales = ['Avoine', 'Blé', 'Céréales Autres', 'Maïs', 'Millet', 'Orge', 'Riz (Eq Blanchi)', 'Seigle', 'Sorgho']
dispo_alimentaire3 = dispo_alimentaire[dispo_alimentaire.Produit.isin(cereales)]
#on calcule la part des céréales utilisée pour l'alimentation humaine et animale notamment
Propc_animale = (dispo_alimentaire3["Aliments pour animaux"].sum() / dispo_alimentaire3["Disponibilité intérieure"].sum())*100
Propc_perdue = (dispo_alimentaire3["Pertes"].sum() / dispo_alimentaire3["Disponibilité intérieure"].sum())*100
Propc_humaine = (dispo_alimentaire3["Nourriture"].sum() / dispo_alimentaire3["Disponibilité intérieure"].sum())*100
Propc_sem = (dispo_alimentaire3["Semences"].sum() / dispo_alimentaire3["Disponibilité intérieure"].sum())*100
Propc_traite = (dispo_alimentaire3["Traitement"].sum() / dispo_alimentaire3["Disponibilité intérieure"].sum())*100
Propc_autres = (dispo_alimentaire3["Autres Utilisations"].sum() / dispo_alimentaire3["Disponibilité intérieure"].sum())*100
print("La proportion de céréales est de", round(Propc_animale), "% pour la part attribuée à l'alimentation animale.")
print("La proportion de céréales est de", round(Propc_perdue), "% pour la part qui est perdue.")
print("La proportion de céréales est de", round(Propc_humaine), "% pour la part attribuée à l'alimentation humaine.")
print("La proportion de céréales est de", round(Propc_sem), "% pour la part attribuée à la semence.")
print("La proportion de céréales est de", round(Propc_traite), "% pour la part attribuée au traitement.")
print("La proportion de céréales est de", round(Propc_autres), "% pour la part attribuée aux autres utilisations.")

```

```python
dispo_alimentaire3c = dispo_alimentaire3.groupby(['Produit'], as_index = False) #On regroupe par produit
dispo_alimentaire3c = dispo_alimentaire3c.sum()
#On ne garde que les colonnes utiles
dispo_alimentaire3c = dispo_alimentaire3c [['Produit','Aliments pour animaux', 'Autres Utilisations', 'Disponibilité intérieure', 'Nourriture', 'Pertes','Semences', 'Traitement']]
#On calcule les proportions de céréales mais cette fois par type de céréales
dispo_alimentaire3c['Proportion animale/produit'] = (dispo_alimentaire3c['Aliments pour animaux']/dispo_alimentaire3c['Disponibilité intérieure'])*100
dispo_alimentaire3c['Proportion perdue/produit'] = (dispo_alimentaire3c['Pertes']/dispo_alimentaire3c['Disponibilité intérieure'])*100
dispo_alimentaire3c['Proportion alimentation humaine/produit'] = (dispo_alimentaire3c['Nourriture']/dispo_alimentaire3c['Disponibilité intérieure'])*100
dispo_alimentaire3c['Proportion semences/produit'] = (dispo_alimentaire3c['Semences']/dispo_alimentaire3c['Disponibilité intérieure'])*100
dispo_alimentaire3c['Proportion traitement/produit'] = (dispo_alimentaire3c['Traitement']/dispo_alimentaire3c['Disponibilité intérieure'])*100
dispo_alimentaire3c['Proportion autres utilisations/produit'] = (dispo_alimentaire3c['Semences']/dispo_alimentaire3c['Disponibilité intérieure'])*100
dispo_alimentaire3c = dispo_alimentaire3c [['Produit','Proportion animale/produit','Proportion perdue/produit', 'Proportion alimentation humaine/produit', 'Proportion semences/produit', 'Proportion traitement/produit', 'Proportion autres utilisations/produit']]
dispo_alimentaire3c = dispo_alimentaire3c.sort_values(by = 'Proportion alimentation humaine/produit', ascending = False)
dispo_alimentaire3c.head(9)
```
